# Is het verdamping?

In dit experiment is er een verwarmingselement in een met water gevulde maatbeker gestopt. Elke minuut is de temperatuur van het water gemeten. Deze metingen zijn opgeslagen in [tempmetingen.csv](tempmetingen.csv).

Verder is gegeven dat: 
$m_{maatbeker} = 820.8 \mathrm{gr}$  
$m_{maatbeker + water, start} = 1292.9 \mathrm{gr}$  
$m_{maatbeker + water, eind} = 1274.9 \mathrm{gr}$

Het moge duidelijk zijn dat er water is 'verdwenen'. Ook is de eindtemperatuur van het water niet gelijk aan een proces waarbij verdamping en eventuele warmteverliezen niet meegenomen worden.

```{exercise}
Onderzoek het proces door:

2. Bepaal of de totale hoeveelheid toegevoegde energie correspondeert met de opwarming en verdamping van het water, geef daarbij de aannames die je doet.
3. Geef aanbevelingen voor een verbetering van het experiment om (nog) beter het thermodynamische proces te kunnen onderzoeken en beschrijven.
```

````python
import numpy as np
import matplotlib.pyplot as plt
#c = r"Downloads/Studie TN/1e jaar/IP 1/thermolab/Content/Labs/tempmetingen.csv"
data = np.loadtxt(tempmetingen.csv, delimiter=';', skiprows=1)
t = data[1: ,0]
T = data[1: , 1]
# parameters
m = 0.4721        # kg
c_w = 4180        # J/(kg K)
P = 100           # W

# tijd in seconden
t_sec = t * 60

# begintemperatuur uit eerste meting
T0 = T[0]

# model
T_model = T0 + (P / (m * c_w)) * t_sec
plt.plot(t, T, 'k.', label="data")
plt.plot(t, T_model, 'b--', label='verwachte lineaire trendlijn')
plt.legend()
plt.xlabel('$t$ (s)')
plt.ylabel('$T$ (°C)')
plt.xlim(min(t), max(t))
plt.show()
````

````python
# massa's (in gram)
m_maatbeker = 820.8
m_start = 1292.9
m_eind = 1274.9

# temperaturen uit metingen (in °C)
T_start = 21.2 
T_eind = 49.4 

c_water = 4180      # soortelijke warmte van water in J/(kg*K)
L_v = 2.26e6        # verdampingswarmte water in J/kg

# Hierbij is er aangenomen dat het
# verwarmingselement ongeveer 100 W gebruikt.

P = 100             # W 
t_min = 22          # min
t_sec = t_min * 60  # s

# beginmassa van het water (in kg)
m_water = (m_start - m_maatbeker) / 1000

# verdampte massa (in kg)
m_verdamp = (m_start - m_eind) / 1000

# temperatuurverandering (in K)
delta_T = T_eind - T_start

# toegevoegde energie (in J)
Q_toegevoegd = P * t_sec

# energie voor de opwarming van het water (in J)
Q_opwarming = m_water * c_water * delta_T      

# energie die het verdampte water bevat (in J)
Q_verdamping = m_verdamp * L_v               

# totale energie (in J)
Q_nodig = Q_opwarming + Q_verdamping


print("De totale benodigde energie om het water op te warmen, met een verwarmingselement van 100W is:", Q_nodig, "J")

````

De data volgt de trendlijn helemaal niet, omdat er wat water verloren gaat door verdamping. De verdamping kost zoveel energie dat de temperatuur niet ver zal stijgen. Dit betekent dat de verwachte trendlijn, zoals die is geplot, ver boven de datapunten ligt. Een aanbeveling om het experiment beter uit te kunnen voeren, zou kunnen zijn dat de maatbeker beter wordt afgesloten en geïsoleerd. Hierdoor wordt er minder warmte de omgeving in gezonden die niet voor temperatuurstijgingen gebruikt kan worden en wordt het resultaat beter. Daarnaast is het nauwkeurig meten van het vermogen van het verwarmingselement belangrijk, zodat een betere schatting kan worden gemaakt hoeveel energie er aan het water wordt overgedragen en er vervluchtigd.
